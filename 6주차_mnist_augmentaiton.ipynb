```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tqdm import tqdm

# GPU 사용 설정 확인 및 활성화
strategy = tf.distribute.get_strategy()
print("REPLICAS: ", strategy.num_replicas_in_sync)

# MNIST 데이터셋 로드
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 데이터 전처리 및 RGB 변환
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1).astype('float32') / 255
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1).astype('float32') / 255
x_train = tf.convert_to_tensor(x_train)
x_test = tf.convert_to_tensor(x_test)
x_train = tf.image.grayscale_to_rgb(x_train)
x_test = tf.image.grayscale_to_rgb(x_test)


# 모델 정의
def create_model():
    with strategy.scope():
        model = Sequential([
            Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 3)),
            MaxPooling2D((2, 2)),
            Conv2D(64, (3, 3), activation='relu'),
            MaxPooling2D((2, 2)),
            Flatten(),
            Dense(64, activation='relu'),
            Dense(10, activation='softmax')
        ])
        model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# 광학 증폭 조합 정의
augmentations = {
    "Brightness": {"brightness_range": (0.8, 1.2)},
    "Contrast": {"contrast_range": (0.8, 1.2)},
    "Hue": {"hue_range": (-0.1, 0.1)},
    "Saturation": {"saturation_range": (0.8, 1.2)},
    "Combined": {
        "brightness_range": (0.8, 1.2),
        "contrast_range": (0.8, 1.2),
        "hue_range": (-0.1, 0.1),
        "saturation_range": (0.8, 1.2)
    },
    "None": {}  # 증폭 없음
}

# 결과를 저장할 딕셔너리 생성
results = {aug_name: {'accuracy': [], 'loss': []} for aug_name in augmentations.keys()}

def apply_augmentations(image, aug_params):
    image = tf.convert_to_tensor(image)
    if 'brightness_range' in aug_params:
        image = tf.image.random_brightness(image, max_delta=aug_params['brightness_range'][1] - 1)
    if 'contrast_range' in aug_params:
        image = tf.image.random_contrast(image, *aug_params['contrast_range'])
    if 'hue_range' in aug_params:
        image = tf.image.random_hue(image, aug_params['hue_range'][1])
    if 'saturation_range' in aug_params:
        image = tf.image.random_saturation(image, *aug_params['saturation_range'])
    return image

# 각 증폭 조합에 대해 정확도 측정
for aug_name, aug_params in tqdm(augmentations.items(), desc="Augmentation Iterations"):
    datagen = ImageDataGenerator(
        preprocessing_function=lambda x: apply_augmentations(x, aug_params),
        fill_mode='nearest'
    )
    model = create_model()
    history = model.fit(datagen.flow(x_train, y_train, batch_size=32),
                        epochs=5, validation_data=(x_test, y_test), verbose=1)

    # 각 epoch의 결과 저장
    results[aug_name]['accuracy'] = history.history['val_accuracy']
    results[aug_name]['loss'] = history.history['val_loss']

# DataFrame 생성 및 pivot_table 생성
df = pd.DataFrame()
for aug_name, values in results.items():
    for epoch, (acc, loss) in enumerate(zip(values['accuracy'], values['loss']), 1):
        new_row = pd.DataFrame({
            'Augmentation': [aug_name],
            'Epoch': [epoch],
            'Accuracy': [acc],
            'Loss': [loss]
        })
        df = pd.concat([df, new_row], ignore_index=True)

# pivot_table 생성
df_pivot = df.pivot_table(values=['Accuracy', 'Loss'],
                          index='Augmentation',
                          columns='Epoch',
                          aggfunc='first')

# 열 이름 재설정
df_pivot.columns = [f'{col[0]}_{col[1]}' for col in df_pivot.columns]

# 마크다운 형식으로 변환
markdown_table = df_pivot.to_markdown()

# 마크다운 표 출력
print(markdown_table)

# 각 Augmentation 별 Loss 변화 시각화
plt.figure(figsize=(12, 8))
for aug_name, values in results.items():
    plt.plot(values['loss'], label=f'{aug_name} Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss by Augmentation Technique')
plt.legend()
plt.show()

# 각 Augmentation 별 Accuracy 변화 시각화
plt.figure(figsize=(12, 8))
for aug_name, values in results.items():
    plt.plot(values['accuracy'], label=f'{aug_name} Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy by Augmentation Technique')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# 각 증폭 기법에 대한 confusion matrix 생성 및 시각화
for aug_name, aug_params in augmentations.items():
    model = create_model()
    datagen = ImageDataGenerator(
        preprocessing_function=lambda x: apply_augmentations(x, aug_params),
        fill_mode='nearest'
    )
    model.fit(datagen.flow(x_train, y_train, batch_size=32),
              epochs=5, validation_data=(x_test, y_test), verbose=0)

    # 예측 수행
    y_pred = model.predict(x_test)
    y_pred_classes = np.argmax(y_pred, axis=1)

    # Confusion Matrix 계산
    conf_mat = confusion_matrix(y_test, y_pred_classes)

    # Confusion Matrix 시각화
    plt.figure(figsize=(10, 8))
    sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'Confusion Matrix - {aug_name}')
    plt.show()

    # 정확도, 정밀도, 재현율 계산
    accuracy = np.trace(conf_mat) / np.sum(conf_mat)
    precision = np.diag(conf_mat) / np.sum(conf_mat, axis=0)
    recall = np.diag(conf_mat) / np.sum(conf_mat, axis=1)

    print(f'Augmentation: {aug_name}')
    print(f'Accuracy: {accuracy:.4f}')
    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    print('-' * 50)
```
